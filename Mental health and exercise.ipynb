```python
dfs = [x for x in df['Is exercise difficult for you? What are the difficulties of not being able to exercise?'].str.split(';')]
bwhat={
    "Haven't established an exercise habit yet":0,
    'No suitable and favorite sports':0,
    'Not in the mood':0,
    'There are other things more important than sports':0,
    "Perfectionism, thinking that if you don't do it well, you simply don't do it.":0,
    "I can't find anyone to exercise with":0,
    'No suitable venue':0,
    'Physically inconvenient':0,
    'Busy and lack of time':0,
    'Painful or boring during exercise':0,
    'Exercise is not necessary':0,
    'Exercise has no difficulty for me':0,
    'others':0}
other = []
for item in dfs:
    for select in item:
        if select == "Haven't established an exercise habit yet":
            bwhat["Haven't established an exercise habit yet"] = bwhat["Haven't established an exercise habit yet"] + 1
        elif select == 'No suitable and favorite sports':
            bwhat['No suitable and favorite sports'] =  bwhat['No suitable and favorite sports'] + 1
        elif select == 'Not in the mood':
            bwhat['Not in the mood'] = bwhat['Not in the mood'] + 1
        elif select == 'There are other things more important than sports':
            bwhat['There are other things more important than sports'] = bwhat['There are other things more important than sports'] + 1
        elif select == "Perfectionism, thinking that if you don't do it well, you simply don't do it":
            bwhat["Perfectionism, thinking that if you don't do it well, you simply don't do it."] = bwhat["Perfectionism, thinking that if you don't do it well, you simply don't do it."] + 1
        elif select == "Perfectionism, thinking that if you don't do it well, you simply don't do it.":
            bwhat["Perfectionism, thinking that if you don't do it well, you simply don't do it."] = bwhat["Perfectionism, thinking that if you don't do it well, you simply don't do it."] + 1
        elif select  == "I can't find anyone to exercise with":
            bwhat["I can't find anyone to exercise with"] = bwhat["I can't find anyone to exercise with"] + 1
        elif select == 'No suitable venue':
            bwhat['No suitable venue'] = bwhat['No suitable venue'] + 1
        elif select  == 'Physically inconvenient':
            bwhat['Physically inconvenient'] = bwhat['Physically inconvenient'] + 1
        elif select == 'Busy and lack of time':
            bwhat['Busy and lack of time'] = bwhat['Busy and lack of time'] +1
        elif select  == 'Painful or boring during exercise':
            bwhat['Painful or boring during exercise'] = bwhat['Painful or boring during exercise']+1
        elif select == 'Exercise is not necessary':
            bwhat['Exercise is not necessary'] = bwhat['Exercise is not necessary'] +1
        elif select == 'Exercise has no difficulty for me':
            bwhat['Exercise has no difficulty for me'] = bwhat['Exercise has no difficulty for me'] +1
        else:
            bwhat['others'] = bwhat['others'] + 1
            other.append(select)
bwhat = dict(sorted(bwhat.items(), key=lambda item: item[1]))
names = list(bwhat.keys())
values = list(bwhat.values())
fig = plt.figure(figsize = (10, 6))
plt.barh(range(len(bwhat)), values, tick_label=names)
plt.xlabel('Count')
plt.ylabel('The difficulties')
plt.title('Is exercise difficult for you? What are the difficulties of not being able to exercise?')
plt.show()
```
